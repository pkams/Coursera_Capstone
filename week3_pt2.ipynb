```python
import pandas as pd
```

```python
# importing the dataframe using pandas
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
df
```

```python
# Exploring the list, take only the first table
df_list = df[0]
df_list
```

```python
# Dropping the Nan
df_list_drop = df_list.dropna()
df_list_drop
```

```python
# Reseting the index
df_tor = df_list_drop.reset_index(drop=True)
df_tor
```

```python
df_tor.columns = ['PostalCode', 'Borough', 'Neighborhood']
```
